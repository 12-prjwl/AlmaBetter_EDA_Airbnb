---

### **using IQR technique**

```python
# 14: Removing the outlier
# writing a outlier function for removing outliers in important columns.
def iqr_technique(DFcolumn):
  Q1 = np.percentile(DFcolumn, 25)
  Q3 = np.percentile(DFcolumn, 75)
  IQR = Q3 - Q1
  lower_range = Q1 - (1.5 * IQR)
  upper_range = Q3 + (1.5 * IQR)                        # interquantile range

  return lower_range,upper_range
```

```python
lower_bound,upper_bound = iqr_technique(airbnb_df['price'])

airbnb_df = airbnb_df[(airbnb_df.price>lower_bound) & (airbnb_df.price<upper_bound)]
```

```python
# so the outliers are removed from price column now check with boxplot and also check shape of new Dataframe!

sns.boxplot(x = airbnb_df['price'])
print(airbnb_df.shape)
```

*   Result: The price distribution now is more condensed and better represents the central tendency of the data. The majority of prices now fall within a range of approximately 50 to 150, with fewer outliers compared to the previous graph. A small number of data points above 300 remain as outliers, but the overall distribution appears more manageable for analysis. This indicates that applying the IQR has successfully reduced the impact of extreme values on the dataset.

**3. Average Price Of Each Neighborhood Group using Point Plot**

```python
# Need to group the Airbnb dataset by neighborhood group and calculate the mean of each group

airbnb_df['price'] = pd.to_numeric(airbnb_df['price'], errors='coerce')# Ensure 'price' is numeric

grouped = airbnb_df.groupby("neighbourhood_group")["price"].mean()

neighbourhood_group_avg_price = grouped.reset_index()
neighbourhood_group_avg_price = round(
    neighbourhood_group_avg_price.rename(columns={"price": "avg_price"}), 2
)

print(neighbourhood_group_avg_price[['neighbourhood_group', 'avg_price']])
```

```python
from statistics import mean #import mean function from the statistics module

sns.pointplot(x = 'neighbourhood_group', y='price', data=airbnb_df, estimator = np.mean)

plt.xlabel('Neighbourhood Group',fontsize=14)
plt.ylabel('Average Price',fontsize=14)
plt.title('Average Price by Neighbourhood Group',fontsize=15)
```

* Result: The average price of an Airbnb listing in New York City varies significantly by neighborhood, with Manhattan having the highest average price at \$146/day and the Bronx the lowest at around \$77/day.  The average price increases as you move from the outer boroughs
(Bronx, Brooklyn, Queens, and Staten Island) toward the city center in Manhattan. Interestingly, Queens and Staten Island have relatively similar
average prices despite being in different parts of the city. This data highlights that the cost of living is generally higher in Manhattan,
likely due to its dense population, commercial importance, and high housing demand in centrally located neighborhoods.
